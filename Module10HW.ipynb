```python
import seaborn as sns
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
```

```python
data = pd.read_csv('iris.data')
data
```

```python
sns.scatterplot(x='sepal length', y='sepal width', data=data, hue='species')
```

```python
sns.catplot(
    x='species', 
    y='petal length',  
    kind='box', 
    data=data,
)

#1) Iris-virginica does not have any outliers
#2) Viriginica and versicolor have a larger petal length than sestosa
```

```python
correlation_matrix = data.corr()
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', vmax=1, vmin=-1, center=0, fmt=".2f", square=True)
plt.title('Correlation Heatmap')
plt.show()
```

```python
max_corr = correlation_matrix.unstack().sort_values(ascending=False)[4]
max_corr
```
